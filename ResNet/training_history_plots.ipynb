```python
import os
import pandas as pd
import matplotlib.pyplot as plt

os.makedirs('plots', exist_ok=True)

pollen = pd.read_csv('pollen_history.csv')
resin  = pd.read_csv('resin_history.csv')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, df, title in zip(axes, [resin, pollen], ['Resin', 'Pollen']):
    ax.plot(df['epoch'], df['train_loss'], label='Train loss', linewidth=2)
    ax.plot(df['epoch'], df['val_loss'],   label='Val loss',   linewidth=2, linestyle='--')
    ax.set_title(f'{title} – Loss vs Epochs')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Loss')
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('plots/loss_curves.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, df, title in zip(axes, [resin, pollen], ['Resin', 'Pollen']):
    ax.plot(df['epoch'], df['map50'], color='green', linewidth=2)
    ax.set_title(f'{title} – mAP@50 vs Epochs')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('mAP@50')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('plots/map50_curves.png', dpi=150, bbox_inches='tight')
plt.show()
```
